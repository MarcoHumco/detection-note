### Count object in video stream 

```python
cap = cv2.VideoCapture(1)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    i = 0
    for item in results.pandas().xyxy[0].name:
        if item == "MetalBox ":
            i += 1
            
    font = cv2.FONT_HERSHEY_SIMPLEX 
    cv2.putText(frame, 
                f"num metalbox: {i}", 
                (5,100), 
                font, 1, 
                (0, 255, 0), 
                3, 
                cv2.LINE_4)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
